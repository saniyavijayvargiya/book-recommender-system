## POPULARITY BASED 

```python
ratings_merged = ratings.merge(books, on = "ISBN")    #merge the ratings and books dataframe on common column isbn
ratings_merged.head()
```

```python
ratings_grouped = ratings_merged.groupby("Book-Title").count()["Book-Rating"].reset_index()    
#group the book titles and count how many ratings each book has

ratings_grouped.rename(columns = {"Book-Rating": "Num-Ratings"}, inplace = True)
#rename book rating column to number of ratings

ratings_grouped.head()
```

```python
ratings_average = ratings_merged.groupby("Book-Title").mean(numeric_only = True)["Book-Rating"].reset_index()
#group the book titles and find the average rating for each book

ratings_average.rename(columns = {"Book-Rating": "Avg-Rating"}, inplace = True)
ratings_average.head()
```

```python
populardf = ratings_grouped.merge(ratings_average, on = "Book-Title")
#merge the dataframes with number of ratings and average rating for each book

populardf.head()
```

```python
populardf = populardf[populardf["Num-Ratings"]>=250].sort_values("Avg-Rating", ascending = False).head(50)

#we only want to consider books which have been rated by a minimum of 250 users
#sort the avg rating for books in descending order so we can get the top 50 books

populardf
```

```python
populardf = populardf.merge(books, on = "Book-Title", suffixes = ("", "_remove")).drop_duplicates("Book-Title")
#book title column was being duplicated causing an error in merging
#hence specify a suffix for the duplicate column to differentiate between the two
#anyways later we are going to select the columns that we need 

populardf = populardf[["Book-Title", "Book-Author", "Num-Ratings", "Avg-Rating", "Image-URL-M"]]
populardf
```

## COLLABORATIVE FILTERING BASED

```python
x = ratings_merged.groupby("User-ID").count()["Book-Rating"] > 200
#we only want users who have rated at least 200 books ie. experienced users

experienced_users = x[x].index    #indexes of users who satisfy the condition
```

```python
filtered_ratings = ratings_merged[ratings_merged["User-ID"].isin(experienced_users)]
#filter ratings to keep only those which have been done by the experienced users 

filtered_ratings
```

```python
y = filtered_ratings.groupby("Book-Title").count()["Book-Rating"] > 50
#we only want books that have been rated by at least 50 users ie. popular books

popular_books = y[y].index    #indexes of books that satisfy the condition
```

```python
final_ratings = filtered_ratings[filtered_ratings["Book-Title"].isin(popular_books)]
#filter the dataframe so that we only have those rows where ratings are done by EXPERIENCED users on POPULAR books

final_ratings
```

```python
finaldf = final_ratings.pivot_table(index = "Book-Title", columns = "User-ID", values = "Book-Rating")
#we want a matrix where row is book title, column is user id, value is rating done by them

finaldf
```

```python
finaldf.fillna(0, inplace = True)
finaldf.head()
```

```python
#we can consider every book as a vector of 810 dimensions
#we can use euclidian distance to find which books are similar

from sklearn.metrics.pairwise import cosine_similarity

similarity_score = cosine_similarity(finaldf)
#gives the euclidian distance of every book from every other book

similarity_score
```

```python
sorted(list(enumerate(similarity_score[0])), key = lambda x: x[1], reverse = True)[1:6]

#in place of 0, give the index of the book as per user
#sort the list containing similarity scores of that book
#enumerate gives index, score so we write the lambda function x[1] so that books are sorted not on index but on score
#in the returned series in descending order, top will be the book itself 
#because a book's score with itself is 1
#hence we need books on 1:6 positions
```

```python
def recommend(book_name):
    index = np.where(finaldf.index == book_name)[0][0]    #gives the index of the parameter book name 
    
    distances = similarity_score[index]
    
    similar_books = sorted(list(enumerate(distances)), key = lambda x: x[1], reverse = True)[1:6]
    
    suggestions = []
    for i in similar_books:
        item = [] 
        temp = books[books["Book-Title"] == finaldf.index[i[0]]]
        temp = temp.drop_duplicates("Book-Title")
       
        item.append(list(temp["Book-Title"].values)[0])
        item.append(list(temp["Book-Author"].values)[0])
        item.append(list(temp["Image-URL-M"].values)[0])
        
        suggestions.append(item)
    return suggestions
```

```python
recommend("1984")
```

```python
recommend("The Notebook")
```
